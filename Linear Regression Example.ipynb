```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('prices.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('Per Captia Income')
plt.scatter(df.Year,df['Per Captia Income'],color='red',marker='+')
```

```python
new_df = df.drop('Per Captia Income',axis='columns')
new_df
```

```python
Income = df['Per Captia Income']
Income
```

```python
reg = linear_model.LinearRegression()
reg.fit(new_df,Income)
```

```python
reg.predict([[2020]])
```
